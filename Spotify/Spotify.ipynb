# Description 

This dataset presents a comprehensive compilation of the most streamed songs on Spotify in 2024. It provides extensive insights into each track's attributes, popularity, and presence on various music platforms, offering a valuable resource for music analysts, enthusiasts, and industry professionals. The dataset includes information such as track name, artist, release date, ISRC, streaming statistics, and presence on platforms like YouTube, TikTok, and more.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# General Guidance:
# UTF-8: Default and most common.
# ISO-8859-1 (latin1): Common fallback, especially for Western European languages.
# Windows-1252: Similar to ISO-8859-1, used in many Windows systems.
# If the file reads successfully after specifying the correct encoding, you can proceed with the data analysis and visualization steps.
# Load the dataset
spotify_data = pd.read_csv("spotify_data.csv", encoding ='ISO-8859-1')
```

# 1. Top 10 Artists by Total Spotify Streams

```python
# spotify_data['Spotify Streams']=spotify_data['Spotify Streams'].str.replace(',','').astype(float)
artist_streams = spotify_data.groupby('Artist')['Spotify Streams'].sum().reset_index()
top10_artists = artist_streams.sort_values(by='Spotify Streams',ascending = False).head(10)
```

```python
top10_artists
```

```python
# plot the top 10 artists
plt.figure(figsize=(10,6))
plt.barh(top10_artists['Artist'], top10_artists['Spotify Streams'], color='red')
plt.xlabel('Total Spotify Streams(billions)')
plt.title('Top 10 Artists by Total Spotify Streams')
plt.show()
```

# 2. Top 10 Most Streamed Tracks on Spotify

```python
# Sort the data by "Spotify Streams" to find the top 10 most streamed tracks
top_tracks = spotify_data[['Track', 'Artist', 'Spotify Streams']].sort_values(by='Spotify Streams', ascending=False).head(10)

# Plot the top 10 most streamed tracks
plt.figure(figsize=(10, 6))
plt.barh(top_tracks['Track'] + ' - ' + top_tracks['Artist'], top_tracks['Spotify Streams'], color='lightcoral')
plt.xlabel('Spotify Streams (billions)')
plt.title('Top 10 Most Streamed Tracks on Spotify')
plt.gca().invert_yaxis()
plt.show()
```

# 3. Comparison Between Spotify Streams and YouTube Views

```python
# Convert the 'YouTube Views' column to strings (in case it contains non-string types)
spotify_data['YouTube Views'] = spotify_data['YouTube Views'].astype(str)
```

```python
# Clean YouTube Views
spotify_data['YouTube Views'] = spotify_data['YouTube Views'].str.replace(',', '').astype(float)

# Plot a scatter plot to compare Spotify Streams and YouTube Views
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Spotify Streams', y='YouTube Views', data=spotify_data, alpha=0.7)
plt.xlabel('Spotify Streams (billions)')
plt.ylabel('YouTube Views (billions)')
plt.title('Spotify Streams vs. YouTube Views')
plt.show()
```

# 4. Correlation Between Release Year and Streaming Numbers

```python
# Clean the 'Spotify Streams' column: remove commas and convert to numeric
spotify_data['Spotify Streams'] = spotify_data['Spotify Streams'].str.replace(',', '').astype(float)
```

```python
# Extract year from the release date
spotify_data['Release Year'] = pd.to_datetime(spotify_data['Release Date']).dt.year

# Calculate average streams by release year
yearly_streams = spotify_data.groupby('Release Year')['Spotify Streams'].mean().reset_index()

# Plot the correlation between release year and average streams
plt.figure(figsize=(10, 6))
sns.lineplot(x='Release Year', y='Spotify Streams', data=yearly_streams, marker='o')
plt.xlabel('Release Year')
plt.ylabel('Average Spotify Streams (billions)')
plt.title('Average Spotify Streams by Release Year')
plt.show()
```

# 5. Tracks with the Highest Playlist Reach on Spotify

```python
# Find the top 10 tracks by playlist reach
top_reach_tracks = spotify_data.sort_values(by='Spotify Playlist Reach', ascending=False).head(10)

# Plot the top 10 tracks by playlist reach
plt.figure(figsize=(10, 6))
plt.barh(top_reach_tracks['Track'] + ' - ' + top_reach_tracks['Artist'], top_reach_tracks['Spotify Playlist Reach'], color='mediumseagreen')
plt.xlabel('Spotify Playlist Reach (millions)')
plt.title('Top 10 Tracks by Spotify Playlist Reach')
plt.gca().invert_yaxis()  # Invert the y-axis to have the highest value on top
plt.show()
```

# 6. Comparison of Explicit vs. Non-Explicit Tracks

```python
# Replace 'Explicit' with the correct column name if different
average_streams_explicit = spotify_data.groupby('Explicit Track')['Spotify Streams'].mean().reset_index()
```

```python
# Group by 'Explicit' and calculate the mean of 'Spotify Streams'
average_streams_explicit = spotify_data.groupby('Explicit Track')['Spotify Streams'].mean().reset_index()

# Plotting the average Spotify streams for explicit vs. non-explicit tracks with hue and legend adjustments
plt.figure(figsize=(8, 6))
sns.barplot(
    x='Explicit Track',
    y='Spotify Streams',
    data=average_streams_explicit,
    palette='viridis',
    hue='Explicit Track',
    dodge=False,
    legend=False
)
plt.xlabel('Explicit Track')
plt.ylabel('Average Spotify Streams (millions)')
plt.title('Average Spotify Streams for Explicit vs. Non-Explicit Tracks')
plt.show()
```

# 7.Seasonal Trends in Streaming

```python
# Extract month from the release date
spotify_data['Release Month'] = pd.to_datetime(spotify_data['Release Date']).dt.month

# Calculate average streams by release month
monthly_streams = spotify_data.groupby('Release Month')['Spotify Streams'].mean().reset_index()

# Plot seasonal trends in streaming
plt.figure(figsize=(10, 6))
sns.lineplot(x='Release Month', y='Spotify Streams', data=monthly_streams, marker='o')
plt.xlabel('Release Month')
plt.ylabel('Average Spotify Streams (billions)')
plt.title('Average Spotify Streams by Release Month')
plt.show()
```

# 8. Artist Consistency Over Years

```python
# Calculate average streams by artist and release year
artist_yearly_streams = spotify_data.groupby(['Artist', 'Release Year'])['Spotify Streams'].mean().reset_index()

# Plot artist consistency (example: top 5 artists with most years in dataset)
top_artists_years = artist_yearly_streams['Artist'].value_counts().head(5).index
artist_consistency = artist_yearly_streams[artist_yearly_streams['Artist'].isin(top_artists_years)]

plt.figure(figsize=(12, 8))
sns.lineplot(x='Release Year', y='Spotify Streams', hue='Artist', data=artist_consistency, marker='o')
plt.xlabel('Release Year')
plt.ylabel('Average Spotify Streams (billions)')
plt.title('Artist Consistency Over Years')
plt.show()
```
